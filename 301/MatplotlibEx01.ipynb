## Line plot of sunspot data

Download the `.txt` data for the "Yearly mean total sunspot number [1700 - now]" from the [SILSO](http://www.sidc.be/silso/datafiles) website. Upload the file to the same directory as this notebook.

```python
import os
assert os.path.isfile('yearssn.dat')
```

Use `np.loadtxt` to read the data into a NumPy array called `data`. Then create two new 1d NumPy arrays named `years` and `ssc` that have the sequence of year and sunspot counts.

```python
# YOUR CODE HERE
data = np.loadtxt('yearssn.dat')
year = np.array([y[0] for y in data][1:])
ssc = np.array([s[1] for s in data][1:])
```

```python
assert len(year)==315
assert year.dtype==np.dtype(float)
assert len(ssc)==315
assert ssc.dtype==np.dtype(float)
```

Make a line plot showing the sunspot count as a function of year.

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
# YOUR CODE HERE
plt.figure(figsize=(30,5))
plt.plot(year, ssc, 'black')
plt.title("Number of Sun Spots per Year", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Sun Spots", fontsize=20)
plt.axis([1695,2025,0,300])
plt.tick_params(length=5, width=2,labelsize=20)
# max_val = max(ssc)
# plt.plot([0,2050],[max_val,max_val], 'k-')
```

Describe the choices you have made in building this visualization and how they make it effective.

I chose to add a large title and large axes labels to clearly indicate the purpose of the graph. I originally added a grid but decided to remove it to minimize chart junk. I changed the line color to black to reduce distraction while also keeping visibility against the white background. I also widened the graph to show a more spread out and "honest" version of the data. I increased the length and widths of the ticks to make them more visible. I changed the ranges on both the x and y axes so that the line would not interfere with the border.

Now make 4 subplots, one for each century in the data set. This approach works well for this dataset as it allows you to maintain mild slopes while limiting the overall width of the visualization. Perform similar customizations as above:

* Customize your plot to follow Tufte's principles of visualizations.
* Adjust the aspect ratio/size so that the steepest slope in your plot is *approximately* 1.
* Customize the box, grid, spines and ticks to match the requirements of this data.

```python
# YOUR CODE HERE
one_year = [y for y in year if y < 1800]
one_ssc = [ssc[i] for i in range(len(one_year))]
two_year = [y for y in year if y < 1900 and y >= 1800]
two_ssc = [ssc[i + len(one_year)] for i in range(len(two_year))]
three_year = [y for y in year if y < 2000 and y >= 1900]
three_ssc = [ssc[i + len(one_year) + len(two_year)] for i in range(len(three_year))]
four_year = [y for y in year if y >= 2000]
four_ssc = [ssc[i + len(one_year) + len(two_year) + len(three_year)] for i in range(len(four_year))]
```

```python
plt.figure(figsize=(30,10))
plt.title("Number of Sun Spots per Year", fontsize=20)

plt.subplot(221)
plt.plot(one_year, one_ssc)
plt.title("Sunspots 1700 - 1799", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Sun Spots", fontsize=20)
plt.tick_params(length=5, width=2,labelsize=20)
plt.axis([1695,1800,0,300])

plt.subplot(222)
plt.plot(two_year, two_ssc)
plt.title("Sunspots 1800 - 1899", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Sun Spots", fontsize=20)
plt.tick_params(length=5, width=2,labelsize=20)
plt.axis([1795,1900,0,300])

plt.subplot(223)
plt.plot(three_year, three_ssc)
plt.title("Sunspots 1900 - 1999", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Sun Spots", fontsize=20)
plt.tick_params(length=5, width=2,labelsize=20)
plt.axis([1895,2000,0,300])

plt.subplot(224)
plt.plot(four_year, four_ssc)
plt.title("Sunspots 2000 - Present", fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Number of Sun Spots", fontsize=20)
plt.tick_params(length=5, width=2,labelsize=20)
plt.axis([2000,2020,0,300])

plt.subplots_adjust(bottom = 1, top = 3)
plt.show()
```
